```python
import pandas as pd
import numpy as np
from ast import literal_eval
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def check_duplicates_str(x):
    l = x.split(' ')
    l = list(set(l))
    return l
def check_duplicates_list(x):
    l = list(set(x))
    return l

def sum_one_user(df):
    df_1 = df.groupby('UID').agg({'History': sum,
                                   'Positive': sum,
                                   'Negative': sum}).reset_index()
#     df_1[['History', 'Positive', 'Negative']] = df_1[['History', 'Positive', 'Negative']].apply(lambda x: check_duplicates_list(x))
    df_1['History'] = df_1['History'].apply(lambda x: check_duplicates_list(x))
    df_1['Positive'] = df_1['Positive'].apply(lambda x: check_duplicates_list(x))
    df_1['Negative'] = df_1['Negative'].apply(lambda x: check_duplicates_list(x))
    return df_1

def user_embeddings(df):
    return embeddingsDF[embeddingsDF['NID'].isin(df['History'].loc[0])]

def cos_sim(cl):
    similarities = cosine_similarity(cl)
#     print(similarities)
    return similarities 
# similarities.sum(axis=0)
```

```python
behaviorDF = pd.read_csv('MINDsmall_train/behaviorDF.csv')
embeddingsDF = pd.read_csv('MINDsmall_train/embeddingsDF.csv')
behaviorDF.head(20)
```

```python
behaviorDF_1 = behaviorDF[behaviorDF['UID'].isin (['U38627'])]
behaviorDF_1 = pd.DataFrame(behaviorDF_1)


behaviorDF_1['History'] = behaviorDF_1['History'].apply(lambda x: literal_eval(x))
behaviorDF_1['Negative'] = behaviorDF_1['Negative'].apply(lambda x: literal_eval(x))
behaviorDF_1['Positive'] = behaviorDF_1['Positive'].apply(lambda x: literal_eval(x))
```

```python
behaviorDF_1 = sum_one_user(behaviorDF_1)
embeddings_user_2 = user_embeddings(behaviorDF_1).reset_index()
del embeddings_user_2['index']
display(embeddings_user_2)
```

```python
# cosine similarity matrix for users' news embeddings
ds = cos_sim(embeddings_user_2.loc[:, embeddings_user_2.columns != 'NID'])

# copy df to track clusters
embeddings_user_cl = embeddings_user_2.copy()
embeddings_user_cl['Cluster'] = np.arange(len(embeddings_user_cl))

merge_history = {}
stack = []
stack_nid = []
ds_merged = []

while len(embeddings_user_2)>1:
    # append first element of df without NID column
    stack.append(embeddings_user_2.loc[:, embeddings_user_2.columns != 'NID'].iloc[[0]])
    # append to another list the respective value of NID column
    stack_nid.append(embeddings_user_2['NID'].iloc[[0]].tolist())
#     print('stack step 0', stack)
#     display('stack_nid step 0', stack_nid)
    while len(stack)>0:
        i = stack[-1]
        i_nid = stack_nid[-1][0]
#         print('i',i)
#         display('i_nid',i_nid)
        J = []
        J_nid = []
        # append the nearest vector to i 
        J.append(embeddings_user_2.iloc[np.where(ds[embeddings_user_2.index[embeddings_user_2['NID'] == i_nid]][0] == max([k for k in ds[embeddings_user_2.index[embeddings_user_2['NID'] == i_nid]][0] if round(k,6) != 1.0]))].loc[:, embeddings_user_2.columns != 'NID'])
        J_nid.append(embeddings_user_2['NID'].iloc[np.where(ds[embeddings_user_2.index[embeddings_user_2['NID'] == i_nid]][0] == max([k for k in ds[embeddings_user_2.index[embeddings_user_2['NID'] == i_nid]][0] if round(k,6) != 1.0]))].tolist())
#         print('J',J)
#         print('J_nid',J_nid)
        merge = False
        if len(stack)>=2:
            j = stack[-2]
            j_nid = stack_nid[-2]
#             print('j',j)
#             print('j_nid',j_nid)
            if j_nid in J_nid:
#                 display(stack[-2:])
                formerge = pd.concat(stack[-2:])
#                 display(formerge)
#                 print(cos_sim(formerge))
                d_merge = round(cos_sim(formerge).sum(axis = 0)[0] - 1, 5)
#                 print(d_merge)

                stack.pop()
                stack.pop()
                stack_nid.pop()
                stack_nid.pop()
#                 print('stack after merge',stack)
#                 print('stack_nid after merge',stack_nid)
                
                # assign the same cluster to the nearest neigbbor for merging
                embeddings_user_cl.loc[embeddings_user_cl['NID'] == j_nid[0], 'Cluster'] = embeddings_user_cl.loc[embeddings_user_cl['NID'] == i_nid]['Cluster'].tolist()
                
                # add new row with merged cluster to df
                embeddings_user_2 = embeddings_user_2.set_index('NID')
#                 display(embeddings_user_2)
#                 print('j_nid[0]', j_nid[0])
#                 print(embeddings_user_2.loc[j_nid[0]].name)
                new_row = (embeddings_user_2.loc[i_nid] + embeddings_user_2.loc[str(j_nid[0])])/2
                new_row.name = embeddings_user_2.loc[i_nid].name + ' ' + embeddings_user_2.loc[j_nid[0]].name
#                 print(new_row.name)
                embeddings_user_2 = embeddings_user_2.append([new_row])
                embeddings_user_2 = embeddings_user_2.reset_index()
                embeddings_user_2 = embeddings_user_2.rename(columns = {'index': 'NID'})
                
                # same for df with cluster info
                embeddings_user_cl = embeddings_user_cl.set_index('NID')
                new_row = (embeddings_user_cl.loc[i_nid] + embeddings_user_cl.loc[str(j_nid[0])])/2
                new_row.name = embeddings_user_cl.loc[i_nid].name + ' ' + embeddings_user_cl.loc[j_nid[0]].name
                embeddings_user_cl = embeddings_user_cl.append([new_row])
                embeddings_user_cl = embeddings_user_cl.reset_index()
                embeddings_user_cl = embeddings_user_cl.rename(columns = {'index': 'NID'})
                embeddings_user_cl.loc[embeddings_user_cl['NID'] == new_row.name, 'Cluster'] = embeddings_user_cl.loc[embeddings_user_cl['NID'] == i_nid]['Cluster'].tolist()
                
                embeddings_user_2 = embeddings_user_2[embeddings_user_2['NID'] != i_nid].reset_index()
                embeddings_user_cl = embeddings_user_cl[embeddings_user_cl['NID'] != i_nid].reset_index()
                del embeddings_user_2['index']
                del embeddings_user_cl['index']
                
                embeddings_user_2 = embeddings_user_2[embeddings_user_2['NID'] != j_nid[0]].reset_index()
                embeddings_user_cl = embeddings_user_cl[embeddings_user_cl['NID'] != j_nid[0]].reset_index()
                del embeddings_user_2['index']
                del embeddings_user_cl['index']
#                 print('df after merge',embeddings_user_2)
#                 print('df with cluster after merge',embeddings_user_cl)
                
                # update merge history
                merge_history[d_merge] = str(new_row.name)
                
#                 display(embeddings_user_2)
                display(embeddings_user_cl)
                print(merge_history)
                
                # create cosine similarity matrix for new df
                ds = cos_sim(embeddings_user_2.loc[:, embeddings_user_2.columns != 'NID'])
#                 print('ds after merge', ds)
                merge = True
            

        if merge == False:
            stack.append(J[0])
            stack_nid.append(J_nid[0])
#             print('stack step 1',stack)
#             print('stack_nid step 1',stack_nid)

print()
for k, v in sorted(merge_history.items(), reverse = True): 
    print(k, v)
# display('final', embeddings_user_cl)
```
